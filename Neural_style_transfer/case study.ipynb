```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = [
    "I love the camera quality.",
    "The battery life is too short",
    "Amazing phone with great features",
    "Screen size too small"
]
```

```python
vectorizer = CountVectorizer()
```

```python
X_bow = vectorizer.fit_transform(corpus)
```

```python
print("BoW Feature Names:", vectorizer.get_feature_names_out())
print("BoW Matrix: \n", X_bow.toarray())
```

```python
tfidf = TfidfVectorizer()
```

```python
X_tfidf = tfidf.fit_transform(corpus)
print("TF-IDF Feature Names:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:", X_tfidf.toarray())
```

```python
import nltk

# Correct resource names
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

sentence = "The quick brown fox jumps over the lazy dog."

# Tokenize the sentence
tokens = nltk.word_tokenize(sentence)

# Get POS tags
pos_tags = nltk.pos_tag(tokens)

# Print the result
print("POS Tags:", pos_tags)
```

```python
grammar = "NP: {<DT>?<JJ>*<NN>}"

chunk_parser = nltk.RegexParser(grammar)
tree = chunk_parser.parse(pos_tags)

print("Noun Phrase Chunks:")
for subtree in tree.subtrees(filter=lambda t: t.label() == 'NP'):
    print(" ".join(Word for word, tag in subtree.leaves()))
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp("The quick brown fox jumps over the lazy dog.")

print("Depnedency Parse:")
for token in doc:
    print(f"{token.text:<10} {token.dep_:<10} --> {token.head.text}")
```
